__Time to split the iris dataset into Training:Tesing datset. Remember there is no standard approach fro this    dividation even though we divide, Based on suggestions from ML/Data science leaders 70:30 approach is good.__

```python
#Import train_test_split
from sklearn.model_selection import train_test_split
```

```python
#Split iris dataset into training and testing datset
trainIris , testIris = train_test_split(iris,test_size = 0.3)
```

```python
#Look over training set
trainIris.head()
```

```python
#Look over testing set
testIris.head()
```

__Testing dataset must not hold the target variable/outcomes, so that we can predict the outcome 
using our trained regression model from trainig datset__ 

```python
#Drop Species from testing dataset
testIris = testIris.drop(['species'],axis=1)
```
